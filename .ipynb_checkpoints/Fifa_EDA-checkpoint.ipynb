# Top Highest paid players

```python
# top highest paid players = TPP

TPP = df[['short_name', 'wage_eur', 'club_name']].sort_values('wage_eur', ascending=False).head(10)
```

```python
TPP
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
TPP.plot(kind = 'bar', title = 'Top Highest paid players at 2011', x = 'short_name')
```

### Messi was the highest paid player as of 2021

# Player rating

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['overall'], bins=30, kde=True)
plt.title('Distribution of Player Ratings')
plt.xlabel('Overall Rating')
plt.ylabel('Frequency')
plt.show()
```

shows that most players are average with only a few percentage of players with high ratings

```python
over_90 = df[df['overall'] >= 90][['short_name', 'overall']].sort_values('overall', ascending=False)
```

Players with rating from 90 and above

```python
over_90
```

### Club with high rated players

```python
top_clubs = df[df['overall'] > 85]['club_name'].value_counts()
top_clubs.plot(kind='bar', color='skyblue', figsize=(15, 6))
plt.title('Top Clubs with Most Players Rated Over 85')
plt.xlabel('Club')
plt.ylabel('Number of Players')
plt.xticks(rotation=45)
plt.show()
```

Liverpool and Real Madrid have the highest number of high rated players
