# Player rating

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['overall'], bins=30, kde=True)
plt.title('Distribution of Player Ratings')
plt.xlabel('Overall Rating')
plt.ylabel('Frequency')
plt.show()
```

shows that most players are average with only a few percentage of players with high ratings

```python
over_90 = df[df['overall'] >= 90][['short_name', 'overall']].sort_values('overall', ascending=False)
```

Players with rating from 90 and above

```python
over_90
```

### Club with high rated players

```python
top_clubs = df[df['overall'] > 85]['club_name'].value_counts()
top_clubs.plot(kind='bar', color='skyblue', figsize=(15, 6))
plt.title('Top Clubs with Most Players Rated Over 85')
plt.xlabel('Club')
plt.ylabel('Number of Players')
plt.xticks(rotation=45)
plt.show()
```

Liverpool and Real Madrid have the highest number of high rated players

## Height distribution by position

```python
main_positions = ['GK', 'CB', 'LB', 'RB', 'CM', 'CDM', 'CAM', 'LW', 'RW', 'ST'] 
df_main = df[df['team_position'].isin(main_positions)]
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='team_position', y='height_cm', data=df_main)
plt.xticks(rotation=90)  # Rotate position labels for better readability
plt.title("Distribution of Player Height Across Positions")
plt.show()
```

The tallest players are usually the goal keepers

## Skills correlation

```python
skills_df = df[['pace', 'shooting', 'passing', 'dribbling', 'defending', 'physic']]
```

```python
plt.figure(figsize=(10, 6))
corr_matrix = skills_df.corr()

sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Between Football Skills")
plt.show()
```

```python
skills = ['pace', 'shooting', 'passing', 'dribbling', 'defending', 'physic']
main_positions = ['CB', 'LB', 'RB', 'CM', 'CDM', 'CAM', 'LW', 'RW', 'ST'] 

for pos in main_positions:
    plt.figure(figsize=(8, 5))
    subset = df_main[df_main['team_position'] == pos][skills]  # Filter by position
    corr_matrix = subset.corr()

    sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
    plt.title(f"Skill Correlation for {pos}")
    plt.show()
```
